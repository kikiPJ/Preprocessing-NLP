```python
file = open("text artikel 2.txt", "r")
f = file.read()
print(f)
```

```python
def sentence_tokenization(s):
    sentences_list = sent_tokenize(s)
    
    return sentences_list

text_data = f
sentence_tokenization(text_data)
```

```python
def word_tokenization(s):
    tokens = word_tokenize(s)

    return tokens
    
text_data = f
word_tokenization(text_data)
```

```python
def casefolding(s):
    new_str = s.lower()
    
    return new_str

text_data = f
casefolding(text_data)
```

```python
def enterNormalize(self, text):
    norm_enter = text.replace("\n", " ")
    return norm_enter
```

```python
text_data = f
text_enterNormalize = enterNormalize('text',text_data)
text_enterNormalize
```

```python
def lowerNormalize(self, text):
        norm_lower = text.lower()
        return norm_lower
```

```python
text_lower = lowerNormalize('text',text_enterNormalize)
text_lower
```
